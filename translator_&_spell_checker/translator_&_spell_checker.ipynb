```python
import pandas as pd
import numpy as np
import re
from deep_translator import GoogleTranslator
```

* We can clear see that the dataset has nearly 1,23,100 unique data out of 175621.  
* Progess that much Data will take More compile time.  
* so I am going to take 1000 data randomly and do the Process.. 

```python
df1 = df.drop_duplicates()
df1.head()
```

```python
df1['clean_words'] = [re.sub(r'[^\w\s]', '', i) for i in list(df1['English words/sentences'])]
df1.head()
```

```python
# shuffle the DataFrame rows
df1 = df1.sample(frac = 1)
df1 = df1.head(1000)
df1.head()
```

```python
df1['To_french'] = df1['clean_words'].apply(lambda x : GoogleTranslator(source='auto', target='fr').translate(x))
df1.head()
```

```python
df1['To_spanish'] = df1['clean_words'].apply(lambda x : GoogleTranslator(source='auto', target='es').translate(x))
df1.head()
```

```python
df1['To_german'] = df1['clean_words'].apply(lambda x : GoogleTranslator(source='auto', target='de').translate(x))
df1.head()
```

# **TASK 02 : Spell Checker**

## Create a program that needs to automatically correct that spelling from the word or a given sentence. (Language : English)

**Note:  show 100 data point outputs both in words and sentence.**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_fwf("/content/spell-testset2.txt")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#splitting the appeal: apeal	 column into Correct_word and Misspelled_word
df[["Correct_word", "Misspelled_word"]]  = df['appeal: apeal'].str.split(" ",n=1, expand = True)
df.head()
```

```python
from textblob import TextBlob
```

```python
df["After_Correction"]=df['Misspelled_word'].apply(lambda x: str(TextBlob(x.split()[0]).correct()))
```

```python
df.head(15)
```

```python
s = [1 for i,j in zip(df['Correct_word'],df['After_Correction']) if i[:len(i)-1] == j]
print("out of 362, "+str(sum(s))+" were predicted exactly to the original word")
```
